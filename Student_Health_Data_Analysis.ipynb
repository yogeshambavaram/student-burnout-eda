```python
import matplotlib.pyplot as plt
import seaborn as sns

# First, we tell Python to only look at columns with numbers (ignoring text columns like 'gender')
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the mathematical correlation between every single variable
corr_matrix = numeric_df.corr()

# Create a massive, beautiful heatmap to visualize it
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)

plt.title('What Actually Causes Student Burnout?', fontsize=18)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set chart style
sns.set_theme(style="whitegrid")

# Create a 2D Histogram (Heatmap) to handle massive amounts of data
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='screen_time_hours', y='daily_sleep_hours', bins=20, cmap='mako', cbar=True)

plt.title('Concentration of Screen Time vs. Sleep', fontsize=16)
plt.xlabel('Daily Screen Time (Hours)', fontsize=12)
plt.ylabel('Daily Sleep (Hours)', fontsize=12)

# Show the plot
plt.show()
```
